## Audios Segmentation

```python
""""
This script segments the audio in 3 sec chunks, for both origianal and spoofed audio. 
These chunk are then furether coonverted into spectrogram images.
"""

from pydub import AudioSegment
import os
import math

def split_audio_folder(input_dir, output_dir, start_idx, chunk_length, chunk_prefix="fake_chunk"):

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Initialize chunk index
    chunk_index = start_idx

    # Process files in sequential order
    for file_name in sorted(os.listdir(input_dir)):
        if file_name.endswith(".wav"):  # Process only .wav files (adjust as needed)
            file_path = os.path.join(input_dir, file_name)
            print(f"Processing file: {file_path}")

            # Load the audio file
            audio = AudioSegment.from_file(file_path)

            # Calculate the number of chunks
            total_length = len(audio) / 1000  # Convert to seconds
            num_chunks = math.ceil(total_length / chunk_length)

            # Split and save each chunk
            for i in range(num_chunks):
                start_time = i * chunk_length * 1000  # Start time in ms
                end_time = min(start_time + chunk_length * 1000, len(audio))  # End time in ms
                chunk = audio[start_time:end_time]

                # Generate the chunk name
                chunk_name = os.path.join(output_dir, f"{chunk_prefix}_{chunk_index}.wav")
                chunk.export(chunk_name, format="wav")
                print(f"Exported {chunk_name}")

                chunk_index += 1  # Increment the chunk index
```

## Spectrogram of Audio Clips

```python
"""
This script generates spectrogram of the audio files using STFT. It loads the audio files, compute STFT. 
It compute STFT with specified windows size and hop length. The resulting spectrogram image provides a visual 
representation of the frequency content of the audio signal over time.

"""
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def generate_spectrogram(audio_file, output_path, cmap='viridis', output_name="chunk"):
    try:
        # Load the audio file
        audio, fs = librosa.load(audio_file, sr=None)  # Preserve original sampling rate

        # Compute the spectrogram
        S = librosa.stft(audio, n_fft=2048, hop_length=512)  # Adjust n_fft and hop_length as needed
        S_db = librosa.amplitude_to_db(np.abs(S), ref=np.max)

        # Plot the spectrogram
        plt.figure(figsize=(6, 6), dpi=100)
        plt.axis('off')  # Remove axes for a clean image
        librosa.display.specshow(S_db, sr=fs, x_axis=None, y_axis=None, cmap=cmap)

        # Save the spectrogram as an image
        spectrogram_file = os.path.join(output_path, f"{output_name}.png")
        plt.savefig(spectrogram_file, bbox_inches='tight', pad_inches=0)
        plt.close()
        print(f"Spectrogram saved to: {spectrogram_file}")

    except Exception as e:
        print(f"Error generating spectrogram for {audio_file}: {e}")

def process_audio_folder(folder_path, output_path, start_index, cmap='viridis'):
    try:
        # Ensure the output directory exists
        if not os.path.exists(output_path):
            os.makedirs(output_path)

        # List all .wav files and sort them by name
        files = sorted([f for f in os.listdir(folder_path) if f.endswith('.wav')], key=lambda x: int(x.split('_')[-1].split('.')[0]))

        if not files:
            print("No .wav files found in the folder.")
            return

        # Generate spectrograms for each file in sequential order
        current_index = start_index
        for file in files:
            file_path = os.path.join(folder_path, file)
            output_name = f"fake_chunk_{current_index}"  # Name with sequential index
            generate_spectrogram(file_path, output_path, cmap, output_name)
            current_index += 1

    except Exception as e:
        print(f"Error processing folder {folder_path}: {e}")
```

## Dataset Preparation


```python
"""
This script prepares a dataset by splitting spectrogram images into training, validation, and 
test sets with specified ratios. It organizes the images into separate folders for "original" 
and "fake" classes under each split (train, validation, test).

It split the spectrogram images into 70% train, 20% validation and 10% test.
"""

import os
import shutil
import random

def prepare_dataset(original_dir, fake_dir, output_dir, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    splits = ['train', 'validation', 'test']
    classes = ['original', 'fake']
    for split in splits:
        for cls in classes:
            split_dir = os.path.join(output_dir, split, cls)
            os.makedirs(split_dir, exist_ok=True)
    
    original_images = [f for f in os.listdir(original_dir) if f.endswith('.png')]
    fake_images = [f for f in os.listdir(fake_dir) if f.endswith('.png')]
    
    # Shuffle the images to ensure randomness
    random.shuffle(original_images)
    random.shuffle(fake_images)
    
    # Split the images into train, validation, and test sets
    def split_data(images, split_ratios):
        total_images = len(images)
        train_end = int(total_images * split_ratios[0])
        val_end = train_end + int(total_images * split_ratios[1])
        train_images = images[:train_end]
        val_images = images[train_end:val_end]
        test_images = images[val_end:]
        return train_images, val_images, test_images
    
    original_train, original_val, original_test = split_data(original_images, [train_ratio, val_ratio, test_ratio])
    fake_train, fake_val, fake_test = split_data(fake_images, [train_ratio, val_ratio, test_ratio])
    
    def copy_files(file_list, source_dir, target_dir):
        for file in file_list:
            shutil.copy(os.path.join(source_dir, file), os.path.join(target_dir, file))
    
    copy_files(original_train, original_dir, os.path.join(output_dir, 'train', 'original'))
    copy_files(original_val, original_dir, os.path.join(output_dir, 'validation', 'original'))
    copy_files(original_test, original_dir, os.path.join(output_dir, 'test', 'original'))
    
    copy_files(fake_train, fake_dir, os.path.join(output_dir, 'train', 'fake'))
    copy_files(fake_val, fake_dir, os.path.join(output_dir, 'validation', 'fake'))
    copy_files(fake_test, fake_dir, os.path.join(output_dir, 'test', 'fake'))
    
    print("Dataset prepared successfully!\n")
    print(f"Training: {len(original_train) + len(fake_train)} spectrogram images")
    print(f"Validation: {len(original_val) + len(fake_val)} spectrogram images")
    print(f"Testing: {len(original_test) + len(fake_test)} spectrogram images")
```

## Model Architecture

```python
"""""
This script defines the model architecture and training parameters. The model is trained on 
GreatLake server utilizing Nvidia's A40 GPU.

The architecture consist of four CNN blocks with one RNN block. The model was trained 
on a batch size of 8 with images resized to 256x256 pixels for 100 epochs.
"""
from tensorflow import keras
from tensorflow.keras import layers, regularizers


def build_model(img_height, img_width):
    input = layers.Input(shape=(img_height, img_width, 3))

    # conv block 1
    x = layers.Conv2D(filters=32, kernel_size=3, activation="relu", kernel_regularizer=regularizers.l2(0.001))(input)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Dropout(0.5)(x)

    # conv block 2
    x = layers.Conv2D(filters=64, kernel_size=3, activation="relu", kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Dropout(0.5)(x)

    # conv block 3
    x = layers.Conv2D(filters=128, kernel_size=3, activation="relu", kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Dropout(0.5)(x)

    # conv block 4
    x = layers.Conv2D(filters=256, kernel_size=3, activation="relu", kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    x = layers.Dropout(0.5)(x)

    # Reshape the data to be compatible with RNNs: (batch_size, time_steps, features)
    x = layers.Reshape((x.shape[1] * x.shape[2], x.shape[3]))(x)

    # RNN layer
    x = layers.LSTM(64, return_sequences=False)(x)
    outputs = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model(inputs=input, outputs=outputs)
    return model


def compile_model(model):
    model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=1e-4),loss='binary_crossentropy',metrics=['accuracy'])

def train_model(model, train_dataset, validation_dataset, model_name, epochs, class_weights=None):
    callbacks = [
        keras.callbacks.ModelCheckpoint(model_name, save_best_only=True, monitor="val_loss"),
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=20, min_lr=1e-6),
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=40, verbose=1),
    ]
    history = model.fit(train_dataset,
                        validation_data=validation_dataset,
                        epochs=epochs,
                        callbacks=callbacks,
                        class_weight=class_weights)

    model.save(model_name)
    return history
```

## Testing on Test Dataset

```python
"""
This script test the trained model on test dataset and plot the confusion matrix. 

The model evaluation metrics are as follow:
1- Accuracy
2- Precision
3- Recall
4- F1 Score
"""

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# Paths
model_path = r"C:\Users\hasan\Desktop\Project\Model\DeepFake_model.keras"
test_dir_fake = r"C:\Users\hasan\Desktop\Project\Dataset3s\test\fake"
test_dir_original = r"C:\Users\hasan\Desktop\Project\Dataset3s\test\original"

# Load the trained model
model = load_model(model_path)

def load_images_from_folder(folder):
  images = []
  labels = []
  for filename in os.listdir(folder):
    filepath = os.path.join(folder, filename)
    if filename.endswith('.jpg') or filename.endswith('.png'):
      image = tf.keras.utils.load_img(filepath, target_size=(256, 256))
      image = tf.keras.utils.img_to_array(image)
      images.append(image)
      labels.append(0 if 'fake' in folder.lower() else 1)  # Assigning labels
  return images, labels

fake_images, fake_labels = load_images_from_folder(test_dir_fake)
original_images, original_labels = load_images_from_folder(test_dir_original)
images = np.array(fake_images + original_images)
labels = np.array(fake_labels + original_labels)
images = images / 255.0  
predictions = model.predict(images)
predicted_labels = (predictions > 0.5).astype(int).flatten()

cm = confusion_matrix(labels, predicted_labels)

# Calculate accuracy, precision, recall, and F1 score
accuracy = accuracy_score(labels, predicted_labels)
precision = precision_score(labels, predicted_labels)
recall = recall_score(labels, predicted_labels)
f1 = f1_score(labels, predicted_labels)

print(f"\nModel Performance:")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision (Positive Predictive Value): {precision:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
print(f"F1 Score: {f1:.4f}\n")

# Plot confusion matrix
class_labels = ['Spoofed', 'Original']
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

## Testing on New Samples

```python
"""
This script randomly selects 7 to 10 image samples from a specified folder containing spectrogram images of new data
which is utilised during training process. It uses a pre-trained deep learning model to classify each image 
as either "Fake" (label 0) or "Original" (label 1).

Steps:
1. Randomly select 6 to 10 images from the folder.
2. Preprocess the images (resize and normalize).
3. Pass the images through the trained model for predictions.
4. Identify the files either as Spoofed (label 0) or Original (label 1).
"""

import os
import numpy as np
import tensorflow as tf 
from tensorflow.keras.models import load_model
import random

model_path = r"C:\Users\hasan\Desktop\Project\Model\DeepFake_model.keras"
sample_dir = r"C:\Users\hasan\Desktop\Project\Sample"
model = load_model(model_path)

def load_random_samples(folder, num_samples=7):
    images = []
    filenames = []
    all_files = [f for f in os.listdir(folder) if f.endswith('.jpg') or f.endswith('.png')]
    sampled_files = random.sample(all_files, min(num_samples, len(all_files)))
    
    for filename in sampled_files:
        filepath = os.path.join(folder, filename)
        image = tf.keras.utils.load_img(filepath, target_size=(256, 256))
        image = tf.keras.utils.img_to_array(image)
        images.append(image)
        filenames.append(filename)
    return np.array(images), filenames

num_samples = random.randint(6, 10)
sampled_images, sampled_filenames = load_random_samples(sample_dir, num_samples)
sampled_images = sampled_images / 255.0
predictions = model.predict(sampled_images)
predicted_labels = (predictions > 0.5).astype(int).flatten()

print(f"Randomly selected {len(sampled_filenames)} samples from New Samples.\n")
for filename, prediction in zip(sampled_filenames, predicted_labels):
    label = '\nDetected as Spoofed Audio\n' if prediction == 0 else '\nDetected as Original Audio\n'
    print(f"{filename}: {label}")
```
